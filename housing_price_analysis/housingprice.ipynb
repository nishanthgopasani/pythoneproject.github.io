```python
IQR=Q3-Q1
```

```python
LB=Q1-1.5*IQR
up=Q1+1.5*IQR
```

```python
if up<sales.loc['max'] and LB>sales.loc['min']:
    print('outlier on both sides')
elif up<sales.loc['max']:
    print('right')
else: 
    print('left')
```
